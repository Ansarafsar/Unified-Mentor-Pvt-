## Define Functions for ETL and Analysis

```python
# Function to extract data

def extract_data(file_path: str) -> pd.DataFrame:
    return pd.read_csv(file_path)
```

```python
# Function to transform data

def transform_data(df: pd.DataFrame) -> pd.DataFrame:
    # Handle missing values
    # Data type conversions if necessary

    df = df.dropna()
    df['age'] = df['age'].astype(int)
    df['sex'] = df['sex'].map({0: 'female', 1: 'male'})
    return df
```

```python
# Function to load data (for this example, just return the transformed data)

def load_data(df: pd.DataFrame) -> pd.DataFrame:
    return df
```

## Exploratory Data Analysis (EDA)


```python
# Function to perform EDA
def perform_eda(df: pd.DataFrame):
    print("Data Head:\n", df.head())
    print("\n***************************************************************************************\n")
    print("Data Info:\n", df.info())
    print("\n***************************************************************************************\n")
    print("Data Description:\n", df.describe())
    print("\n***************************************************************************************\n")
    print("Missing Values:\n", df.isna().sum())
    
    plt.figure(figsize=(12, 8))
    sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
    plt.title('Correlation Matrix')
    plt.show()
    
    plt.figure(figsize=(10, 6))
    sns.histplot(df['age'], bins=30, kde=True)
    plt.title('Age Distribution')
    plt.show()
    
```

## Visualization


```python
# Function to create visualizations
def create_visualizations(df: pd.DataFrame):
    # Heart disease by gender
    plt.figure(figsize=(8, 6))
    sns.countplot(x='sex', data=df, palette='viridis')
    plt.title('Heart Disease by Gender')
    plt.show()
    
    # Heart disease by age
    plt.figure(figsize=(10, 6))
    sns.histplot(df['age'], kde=True, bins=30)
    plt.title('Heart Disease by Age')
    plt.show()
```

## Key Metrics and Relationships


```python
def find_key_metrics(df: pd.DataFrame):
    print("Average Age of Patients:", df['age'].mean())
    print("Gender Distribution:\n", df['sex'].value_counts(normalize=True))

    # Example: Compare age and cholesterol levels
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x='age', y='chol', data=df, hue='sex')
    plt.title('Age vs Cholesterol Levels')
    plt.show()
    
```

## Main Function to Run ETL and Analysis


```python
def main(file_path: str):
    df = extract_data(file_path)
    df = transform_data(df)
    df = load_data(df)
    perform_eda(df)
    create_visualizations(df)
    find_key_metrics(df)

# Run the main function with the provided dataset
file_path = "C:/Unified Mentor Private Limited/Heart Disease data/Heart Disease data.csv"
main(file_path)
```
