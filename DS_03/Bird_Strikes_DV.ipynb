# Visuals Depicting the Number of Bird Strikes

```python
# Number of bird strikes per year
df['FlightDate'] = pd.to_datetime(df['FlightDate']).dt.year
bird_strikes_per_year = df['FlightDate'].value_counts().sort_index()

# Plot the number of bird strikes per year
plt.figure(figsize=(10, 6))
bird_strikes_per_year.plot(kind='bar', color='skyblue')
plt.title('Number of Bird Strikes per Year')
plt.xlabel('Year')
plt.ylabel('Number of Bird Strikes')
plt.show()
```

# Yearly Analysis & Bird Strikes in the US

```python
# Yearly analysis of bird strikes
yearly_analysis = df.groupby('FlightDate').size()

# Plot yearly analysis of bird strikes
plt.figure(figsize=(12, 6))
yearly_analysis.plot(kind='line', marker='o', linestyle='-', color='b')
plt.title('Yearly Analysis of Bird Strikes in the US')
plt.xlabel('Year')
plt.ylabel('Number of Bird Strikes')
plt.grid(True)
plt.show()
```

# Top 10 US Airlines in terms of having encountered bird strikes

```python
# Top 10 US Airlines with most bird strikes
top_airlines = df['Aircraft: Airline/Operator'].value_counts().head(10)

# Plot top 10 airlines with most bird strikes
plt.figure(figsize=(12, 6))
top_airlines.plot(kind='bar', color='green')
plt.title('Top 10 US Airlines with Most Bird Strikes')
plt.xlabel('Airlines')
plt.ylabel('Number of Bird Strikes')
plt.xticks(rotation=90)
plt.show()
```

# Airports with most incidents of bird strikes – Top 50

```python
# Top 50 airports with most bird strikes
top_airports = df['Airport: Name'].value_counts().head(50)

# Plot top 50 airports with most bird strikes
plt.figure(figsize=(12, 12))
top_airports.plot(kind='barh', color='orange')
plt.title('Top 50 Airports with Most Bird Strikes')
plt.xlabel('Number of Bird Strikes')
plt.ylabel('Airports')
plt.show()
```

# Yearly Cost Incurred due to Bird Strikes

```python
# Yearly cost incurred due to bird strikes
df['Cost: Total $'] = df['Cost: Total $'].replace('[\$,]', '', regex=True).astype(float)
yearly_cost = df.groupby('FlightDate')['Cost: Total $'].sum()

# Plot yearly cost incurred due to bird strikes
plt.figure(figsize=(12, 6))
yearly_cost.plot(kind='bar', color='red')
plt.title('Yearly Cost Incurred due to Bird Strikes')
plt.xlabel('Year')
plt.ylabel('Total Cost ($)')
plt.show()
```

# When do most bird strikes occur?

# Altitude of aeroplanes at the time of strike

```python
# Altitude at the time of bird strikes
plt.figure(figsize=(10, 6))
sns.histplot(df['Altitude bin'], bins=30, kde=True, color='brown')
plt.title('Altitude of Aeroplanes at the Time of Bird Strikes')
plt.xlabel('Altitude (ft)')
plt.ylabel('Frequency')
plt.show()
```

# Phase of flight at the time of the strike

```python
# Phase of flight at the time of the strike
phase_of_flight = df['When: Phase of flight'].value_counts()

# Plot phase of flight at the time of the strike
plt.figure(figsize=(12, 6))
phase_of_flight.plot(kind='bar', color='cyan')
plt.title('Phase of Flight at the Time of Bird Strikes')
plt.xlabel('Phase of Flight')
plt.ylabel('Number of Bird Strikes')
plt.xticks(rotation=45)
plt.show()
```

# Average Altitude of the aeroplanes in different phases at the time of strike

```python
# Define a function to convert 'Altitude bin' to numeric values
def convert_altitude_bin(altitude_bin):
    if pd.isna(altitude_bin):
        return np.nan
    elif '< 1000 ft' in altitude_bin:
        return 500
    elif '> 1000 ft' in altitude_bin:
        return 1500
    else:
        return np.nan

# Apply the conversion function to the 'Altitude bin' column
df['Altitude'] = df['Altitude bin'].apply(convert_altitude_bin)

# Group by 'When: Phase of flight' and calculate the average altitude
avg_altitude_phase = df.groupby('When: Phase of flight')['Altitude'].mean().sort_values()

# Plot the average altitude during different phases of flight
plt.figure(figsize=(12, 6))
avg_altitude_phase.plot(kind='bar', color='magenta')
plt.title('Average Altitude of Aeroplanes in Different Phases at the Time of Bird Strikes')
plt.xlabel('Flight Phase')
plt.ylabel('Average Altitude (ft)')
plt.xticks(rotation=45)
plt.show()
```

# Effect of Bird Strikes & Impact on Flight

```python
# Effect and impact of bird strikes on flights
impact_of_strikes = df['Effect: Indicated Damage'].value_counts()

# Plot effect and impact of bird strikes on flights
plt.figure(figsize=(12, 6))
impact_of_strikes.plot(kind='bar', color='grey')
plt.title('Effect and Impact of Bird Strikes on Flights')
plt.xlabel('Impact')
plt.ylabel('Number of Bird Strikes')
plt.xticks(rotation=45)
plt.show()
```

# Effect of Strike at Different Altitude

```python
# Effect of strikes at different altitudes
# Apply the conversion function to the 'Altitude bin' column
df['Altitude'] = df['Altitude bin'].apply(convert_altitude_bin)

# Group by 'Effect: Indicated Damage' and calculate the average altitude
effect_altitude = df.groupby('Effect: Indicated Damage')['Altitude'].mean().sort_values()

# Plot effect of strikes at different altitudes
plt.figure(figsize=(12, 6))
effect_altitude.plot(kind='bar', color='blue')
plt.title('Effect of Bird Strikes at Different Altitudes')
plt.xlabel('Effect')
plt.ylabel('Average Altitude (ft)')
plt.xticks(rotation=45)
plt.show()
```
